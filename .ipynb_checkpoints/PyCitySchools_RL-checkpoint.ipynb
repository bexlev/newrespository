```python
# Dependencies and Setup
import pandas as pd

# File to Load (using absolute path)
school_data = pd.read_csv(r"C:\Users\Rebecca Levine\Desktop\panda-challenge\Resources\schools_complete.csv")
student_data = pd.read_csv(r"C:\Users\Rebecca Levine\Desktop\panda-challenge\Resources\students_complete.csv")

                              
# # Read School and Student Data File and store into Pandas DataFrames
# school_data = pd.read_csv(school_data_csv)
# student_data = pd.read_csv(student_data_csv)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete.head()
```

# District Summary


```python
# Calculate the total number of schools

number_of_schools = len(school_data_complete["school_name"].unique())
print(number_of_schools)

# Calculate the total number of students
number_of_students = len(school_data_complete["Student ID"])
print(number_of_students)

# Calculate the total budget
#Want to extract only school names and budgets, and find each school's budget. Then they can be summed.
school_budgets = (school_data_complete["school_name"].unique(), school_data_complete["budget"].unique())
print(school_budgets)
#Was possible to visually verify that the budgets are unique across schools so they can just be summed


total_budget = (school_data_complete["budget"].unique()).sum()
print(total_budget)
```

```python
# Calculate the average math score
average_math_score = (school_data_complete["math_score"].mean())
print(average_math_score)

# Calculate the average reading score
average_reading_score = (school_data_complete["reading_score"].mean())
print(average_reading_score)
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)


students_passing_math  = school_data_complete.loc[(school_data_complete["math_score"]>= 70), :]
(len(students_passing_math))
print(number_of_students)

#From earlier calculation, total students is number_of_students
#pp will now stand for percent passing
pp_math = ((len(students_passing_math))/number_of_students)*100
pp_math

rounded_pp_math = round(pp_math)
print(rounded_pp_math)


# Calculate the percentage of students with a passing reading score (70 or greater)

students_passing_reading  = school_data_complete.loc[(school_data_complete["reading_score"]>= 70), :]
print(len(students_passing_reading))

pp_reading = ((len(students_passing_reading))/number_of_students)*100
pp_reading

rounded_pp_reading = round(pp_reading)
print(rounded_pp_reading)



# Calculate the percentage of students who passed math and reading (% Overall Passing)
passing_reading_and_math = school_data_complete.loc[(school_data_complete["reading_score"]>= 70) & (school_data_complete["math_score"]>= 70), :]
print(len(passing_reading_and_math))

pp_reading_and_math = ((len(passing_reading_and_math))/number_of_students)*100
pp_reading_and_math
                                    
rounded_pp_reading_and_math = round(pp_reading_and_math)
print(rounded_pp_reading_and_math)
```

# School Summary

```python
school_types = school_data.set_index("school_name")["type"]
school_types
```

```python
school_pop = school_data.set_index("school_name")["size"]
school_pop
```

```python
school_budget = school_data_complete.groupby("school_name").mean()["budget"]
school_budget
```

```python
per_student_budget = school_budget // school_pop
per_student_budget
```

```python
math_score_per_school = school_data_complete.groupby("school_name").mean()["math_score"]
r_math_score_per_school =round(math_score_per_school)
r_math_score_per_school
```

```python
reading_score_per_school = school_data_complete.groupby("school_name").mean()["reading_score"]
r_reading_score_per_school =round(reading_score_per_school)
r_reading_score_per_school
```

# Percent Passing

```python
passing_math_students = school_data_complete[school_data_complete["math_score"]>=70].groupby("school_name").count()['size']
percent_passing_math = round(passing_math_students/school_pop * 100)
print(percent_passing_math)


passing_reading_students = school_data_complete[school_data_complete["reading_score"]>=70].groupby("school_name").count()['size']
percent_passing_reading = round(passing_reading_students/school_pop * 100)
print(percent_passing_reading)

passing_both = school_data_complete[(school_data_complete["reading_score"]>=70) & (school_data_complete["math_score"]>=70)].groupby("school_name").count()['size']
percent_passing_both = round(passing_both/school_pop * 100)
print(percent_passing_both)

```

```python
#Collecting these values above in a Dataframe
per_school_summary = pd.DataFrame(     
    {
        "School Type": school_types,
         "Total Students": school_pop,
         "Total School Budget": school_budget,
        "Budget per Student" : per_student_budget,
        "Average Math Score": r_math_score_per_school,
        "Average Reading Score": r_reading_score_per_school,
        "Percent Passing Math" : percent_passing_math,
        "Percent Passing Reading": percent_passing_reading,
        "Percent Passing Both Math and Reading" : percent_passing_both,
     })
print(per_school_summary)
```

# Overview Table for School Metrics

```python
# Create an overview table that summarizes key metrics about each school
#Format Mapping for Calculations in Dollar Units

per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map('${:,.2f}'.format)
per_school_summary["Per Student Budget"] = per_school_summary["Budget per Student"].map('${:,.2f}'.format)
per_school_summary




```

# Top Performing Schools (By % Overall Passing)

```python
# Sort and display the top five performing schools by % overall passing.

top_performing_schools_df = per_school_summary.sort_values("Percent Passing Both Math and Reading", ascending=True)
top_performing_schools_df.head(5)

```

# Bottom Performing Schools (By % Overall Passing)

```python
# Sort and display the five worst-performing schools by % overall passing.

bottom_performing_schools_df = per_school_summary.sort_values("Percent Passing Both Math and Reading", ascending=False)
bottom_performing_schools_df.head(5)
```

# Math Scores by Grade

```python
# Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th)
#at each school.

nine = school_data_complete[school_data_complete['grade'] == '9th'].groupby('school_name').mean()['math_score']
ten = school_data_complete[school_data_complete['grade'] == '10th'].groupby('school_name').mean()['math_score']
eleven = school_data_complete[school_data_complete['grade'] == '11th'].groupby('school_name').mean()['math_score']
twelve = school_data_complete[school_data_complete['grade'] == '12th'].groupby('school_name').mean()['math_score']

# Combine the series into a dataframe
math_scores_by_grade = pd.DataFrame({
    '9th':nine,
    '10th':ten,
    '11th':eleven,
    '12th':twelve,
})
math_scores_by_grade.index.name = None
math_scores_by_grade


```

# Reading Score by Grade

```python
# Perform the same operations as above for reading scores

nine = school_data_complete[school_data_complete['grade'] == '9th'].groupby('school_name').mean()['reading_score']
ten = school_data_complete[school_data_complete['grade'] == '10th'].groupby('school_name').mean()['reading_score']
eleven = school_data_complete[school_data_complete['grade'] == '11th'].groupby('school_name').mean()['reading_score']
twelve = school_data_complete[school_data_complete['grade'] == '12th'].groupby('school_name').mean()['reading_score']

# Combine the series into a dataframe
reading_scores_by_grade = pd.DataFrame({
    '9th':nine,
    '10th':ten,
    '11th':eleven,
    '12th':twelve,
})
reading_scores_by_grade.index.name = None
reading_scores_by_grade
```

# Scores by School Spending

```python
# Create a table that breaks down school performances based on average Spending Ranges (Per Student). 
#Use 4 reasonable bins to group school spending. Include in the table each of the following:
                            # Average Math Score
                            # Average Reading Score
                            # Percent Passing Math
                            # Percent Passing Reading
                            # Percent Passing Both Math and Reading (Average of the above two)

                    
#binning
spending_bins = [0, 580, 600, 650, 700]
labels = ["<$580", "$580-600", "$600-650", "$650-700"]
school_spending = per_school_summary.copy()
school_spending['Spending Ranges (Per Student)'] = pd.cut(per_student_budget, spending_bins, labels=labels, right=False)

spending_mean = school_spending.groupby('Spending Ranges (Per Student)').mean()

spending_score_math = spending_mean['Average Math Score']
spending_score_reading = spending_mean['Average Reading Score']
spending_passing_math = spending_mean['Percent Passing Math']
spending_passing_reading = spending_mean['Percent Passing Reading']
spending_passing_overall = spending_mean['Percent Passing Both Math and Reading']

#Creating dataframe

spending_scores = pd.DataFrame({
    "Average Math Score": spending_score_math,
    "Average Reading Score": spending_score_reading,
    "Percent Passing Math": spending_passing_math,
    "Percent Passing Reading": spending_passing_reading,
    "Percent Passing Both Math and Reading": spending_passing_overall
})
spending_scores

```

# Scores by School Size

```python
# Perform the same operations as above, based on school size.

#Binning
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_size = per_school_summary.copy()
school_size['School Size'] = pd.cut(school_size['Total Students'], size_bins, labels=labels, right=False)

size_mean = school_size.groupby('School Size').mean()

size_score_math = size_mean['Average Math Score']
size_score_reading = size_mean['Average Reading Score']
size_passing_math = size_mean['Percent Passing Math']
size_passing_reading = size_mean['Percent Passing Reading']
size_passing_overall = size_mean['Percent Passing Both Math and Reading']

#Creating dataframe
size_scores = pd.DataFrame({
    "Average Math Score": size_score_math,
    "Average Reading Score": size_score_reading,
    "% Passing Math": size_passing_math,
    "Percent Passing Reading": size_passing_reading,
    "Percent Passing Both Math and Reading": size_passing_overall
})
size_scores
```

# Scores by School Type

```python
# Perform the same operations as above, based on school type

#Binning

type_mean = school_size.groupby('School Type').mean()

type_score_math = type_mean['Average Math Score']
type_score_reading = type_mean['Average Reading Score']
type_passing_math = type_mean['Percent Passing Math']
type_passing_reading = type_mean['Percent Passing Reading']
type_passing_overall = type_mean['Percent Passing Both Math and Reading']

#Creating dataframe
type_scores = pd.DataFrame({
    "Average Math Score": type_score_math,
    "Average Reading Score": type_score_reading,
    "Percent Passing Math": type_passing_math,
    "Percent Passing Reading": type_passing_reading,
    "Percent Passing Both Math and Reading": type_passing_overall
})
type_scores
```
